```python
df = df_all.iloc[::-1]

def fill_game_list(team, season):

    games = [[] for i in range(len(df.index))]

    i=0
    
    for j in df.index:
    
        if df.loc[j, 'season'] == season:
        
            if df.loc[j, 'awayteam'] == team:
                win_bool = 1*(df.loc[j, 'awayscore'] > df.loc[j, 'homescore'])
                loss_bool = 1*(not win_bool)
                games[i] = [str(df.loc[j, 'date']), win_bool, loss_bool]
                i += 1

            if df.loc[j, 'hometeam'] == team:
                win_bool = 1*(df.loc[j, 'homescore'] > df.loc[j, 'awayscore'])
                loss_bool = 1*(not win_bool)
                games[i] = [str(df.loc[j, 'date']), win_bool, loss_bool]
                i += 1   

    games = [game for game in games if len(game) != 0] # remove all the extras
    return games
```

```python
''' this version takes a list of seasons and/or teams to plot '''

from scipy.interpolate import spline
from itertools import repeat
from matplotlib.dates import MonthLocator, DateFormatter

def add_text(xloc, yloc, text):
    box = dict(boxstyle='square', facecolor='white')
    plt.text(xloc, yloc, text, bbox=box, fontsize=15)

def plot_moving_sum(teams, seasons, ymin=0, ymax=0, label=[]):
    
    colors = sns.color_palette()
    fig = plt.figure(figsize=(14,10))
    
#    for team, season in zip(teams, seasons):
    t = 0
    kplot = {}
    for team in teams:
        kplot[team] = [i+1 for i in range(len(seasons)*len(teams))][t::len(teams)]
        t += 1

    for team in teams:
        k = 0
        for season in seasons[::-1]:
    
            ax = fig.add_subplot(len(seasons), len(teams), kplot[team][k])
    
            team_games = fill_game_list(team=team, season=season)

            game_dates = [i[0].replace('-','/') for i in team_games]
            wins = [i[1] for i in team_games]
            losses = [i[2] for i in team_games]

            # calcualte the moving sum of wins in the past 5 games
            moving_average, moving_average_xpts = [], []
            # game_dates is the full list of dates (not yet datetime objects) to be plotted
            # wins is the full list of wins (1) and losses (0)
            for i, game in enumerate(game_dates):
                if i > 4:
                    moving_average.append(sum(wins[i-5:i]))
                    moving_average_xpts.append(game)

            # add data per game as a bar plot
            xpts = [datetime.datetime.strptime(d,'%Y/%m/%d').date() for d in game_dates]
            ax.bar(xpts, wins, color=colors[k], alpha=0.5)
            plt.xticks(rotation=20)
                
            # discrete moving average
            xpts = [datetime.datetime.strptime(d,'%Y/%m/%d').date() for d in moving_average_xpts]
            ax.plot(xpts, moving_average, '-o', color=colors[k], alpha=0.7,
                    label=team+' '+str(season)[:4]+' / '+str(season)[4:])
            plt.legend(loc='upper left')
    
            if len(label) != 0 and k == label[0]-1:
                add_text(label[1], label[2], label[3])
                label = []
    
            # define date range of curved line for "continuous" moving average
            start = datetime.datetime.strptime(moving_average_xpts[0],'%Y/%m/%d')
            end = datetime.datetime.strptime(moving_average_xpts[-1],'%Y/%m/%d')
            xnew_all_days = [start + datetime.timedelta(days=x) for x in range(0, 1+(end-start).days)]
            xmovingav = [datetime.datetime.strptime(d,'%Y/%m/%d') for d in moving_average_xpts]
            # quad means we can use 4 data points per day to get a more smooth line
            xnew_quad_days = [start + datetime.timedelta(hours=x*6) for x in range(0, 4*(1+(end-start).days))]

            # convert datetime range to values between 0 and 1
            xnew_skewed_0_1 = []
            moving_average_uniform = []
            j=0
            for i in range(len(xnew_all_days)):
                if xnew_all_days[i] in xmovingav:
                    xnew_skewed_0_1.append(i/(len(xnew_quad_days)-1))
                    # -1 in line above allows the first and final
                    # locations to be at 0% and 100%
                    j += 1*(i>0)
    
            # plot the curved fitted line in datetime coordinates
            x_smooth = np.linspace(xnew_skewed_0_1[0], xnew_skewed_0_1[-1], len(xnew_quad_days))
            y_smooth = spline(xnew_skewed_0_1, moving_average, x_smooth)
            ax.plot(xnew_quad_days, y_smooth, '-', color=colors[k], alpha=0.3)

            ymax = max(moving_average)+0.2*(max(moving_average)) if ymax == 0 else ymax
            plt.ylim(0, ymax)
            plt.setp(ax.get_yticklabels()[0], visible=False)    
            plt.setp(ax.get_yticklabels()[-1], visible=False)
#            y_labels = ax.get_yticklabels()
#            y_labels[0] = y_labels[-1] = " "
#            ax.set_yticklabels(y_labels)
            
            xmin = datetime.datetime.strptime(str(season)[:4]+'/10/01','%Y/%m/%d')
            xmax = xmin + datetime.timedelta(days=6.5*365/12)
            plt.xlim(xmin,xmax)
            
            # format the x-ticks
            months = MonthLocator()  # every month
            monthsFmt = DateFormatter('%B')
            ax.xaxis.set_major_locator(months)
            ax.xaxis.set_major_formatter(monthsFmt)
            if k != len(seasons)-1:
                ax.set_xticklabels([])
       
            k += 1

    plt.tight_layout()
    plt.subplots_adjust(hspace=0.05)
```
